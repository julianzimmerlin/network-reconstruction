```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
g = nx.bull_graph()
nx.draw(g)
```

```python
out_address = './bull_vis/'
```

```python
pos = [0,0,0,0,0]
pos[0] = [0, -0.4]
pos[1] = [-0.4, 0]
pos[2] = [0.4, 0]
pos[3] = [-0.8, 0.3]
pos[4] = [0.8, 0.3]
print(pos)
node_colors = list()
node_colors.append(['#1f78b4', '#1f78b4', '#1f78b4', 'red', '#1f78b4'])
node_colors.append(['#1f78b4', 'red', '#1f78b4', 'red', '#1f78b4'])
node_colors.append(['#1f78b4', 'red', 'red', 'red', '#1f78b4'])
node_colors.append(['#1f78b4', 'red', 'red', '#1f78b4', '#1f78b4'])
node_colors.append(['red', 'red', 'red', '#1f78b4', '#1f78b4'])
node_colors.append(['red', '#1f78b4', 'red', '#1f78b4', '#1f78b4'])

for i in range(6):
    plt.clf()
    plt.xlim(-1,1)
    plt.ylim(-0.5,0.4)
    plt.figure(1, figsize=(10,7))
    plt.text(-0.93, -0.45, 't='+str(i), fontsize=15)
    nx.draw_networkx(g, node_size=1200,width=3.0,edge_color='grey',pos=pos,with_labels=False, node_color=node_colors[i])
    #plt.savefig(out_address+str(i)+'.png', dpi=200)
```

```python
plt.clf()
plt.xlim(-1,1)
plt.ylim(-0.5,0.4)
plt.figure(1, figsize=(10,7))
nx.draw_networkx(g, node_size=1200,width=3.0,edge_color='grey',pos=pos,with_labels=True, node_color = 'black', font_color='white')
plt.savefig(out_address+'graph.png', dpi=200)
```
