```python
import networkx as nx
import matplotlib.pyplot as plt
G0 = nx.barabasi_albert_graph(10, 2, seed=0)
G1 = nx.barabasi_albert_graph(20, 2, seed=0)
pos0 = nx.kamada_kawai_layout(G0)
pos1 = nx.kamada_kawai_layout(G1)
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,4),gridspec_kw={'width_ratios':[1,1]},dpi=300)

nx.draw_networkx_nodes(G0, pos0, node_size=50, node_color='w', linewidths=2.5, edgecolors='.2', ax=ax[0])
nx.draw_networkx_edges(G0, pos0, width=1.75, edge_color='.3', alpha=0.5, ax=ax[0])
nx.draw_networkx_nodes(G1, pos1, node_size=50, node_color='w', linewidths=2.5, edgecolors='.2', ax=ax[1])
nx.draw_networkx_edges(G1, pos1, width=1.75, edge_color='.3', alpha=0.5, ax=ax[1])
ax[0].set_axis_off()
ax[1].set_axis_off()
```

```python
fig.savefig('barabasi_vis/ba10_ba20-png', dpi=300)
```
